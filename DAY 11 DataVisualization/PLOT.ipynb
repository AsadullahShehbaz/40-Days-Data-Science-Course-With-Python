```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ship = sns.load_dataset("titanic")
ship.info()
```

```python
ship.isnull().sum()
# Fill missing values in 'age' with the mean age
```

```python
ship.survived.value_counts().plot(kind = 'bar')
```

```python
sns.countplot(x = 'survived', data = ship)
# Set the figure size
```

```python
plt.scatter(x='age',y='fare',data=ship)
```

```python
sns.scatterplot(x='age',y='fare',hue='pclass',data=ship)
```

```python
sns.barplot(x='sex',y='fare',hue='class',data=ship)
```

```python
sns.pairplot(ship, hue='survived', diag_kind='kde')
```

```python
ship.hist()
```

```python
ship.hist(bins = 50, figsize=(9,7),grid=False,edgecolor='green')
```

```python
ship.hist(
    bins=20,                # More detailed bins
    figsize=(12, 6),        # Larger figure
    grid=True,              # Show grid
    color='green',          # Bar color
    alpha=0.7,              # Semi-transparent
    edgecolor='black'       # Black bar borders
)
# plt.title("Distribution of Ship Data")  # Add title
# plt.xlabel("Values")                   # X-axis label
# plt.ylabel("Frequency")                # Y-axis label
# plt.show()
```

```python
sns.boxplot(x='who',y='fare',hue='who',data=ship)
```

```python
sns.boxenplot(x='sex',y='fare',hue='who',data=ship)
```

```python
g = sns.FacetGrid(ship, hue = 'survived',col="who",margin_titles=True,palette={1:'seagreen',0:'gray'})
g.map_dataframe(plt.scatter, 'age', 'fare' ,edgecolor='w' ).add_legend()
```

```python
sns.stripplot(x='sex',y='fare',hue='survived',data=ship,jitter=True)
```

```python
tip = sns.load_dataset("tips")
sns.stripplot(x='day',y='total_bill',hue='sex',data=tip)
```

```python
sns.violinplot(x='sex',y='fare',hue='class',data=ship)
```

```python
# Selecting only numeric columns for correlation
titanic_numeric = ship[["age", "fare", "pclass", "sibsp", "parch", "survived"]]
corr = titanic_numeric.corr()
# print(correlation_matrix)
```

```python
# Create a heatmap
plt.figure(figsize=(8,6))  # Set figure size
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5, mask=corr < 0.2)

# Show the plot
plt.show()
```
