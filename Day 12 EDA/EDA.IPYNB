```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df2= df 
```

```python
print(df['age'].min())
print(df['age'].max())
print(df['age'].mean())
```

```python
df2['age'].fillna(df2['age'].mean(), inplace=True)
df2
```

```python
df3 = df2 
```

```python
df3 = df3.drop(columns=['deck'])
```

```python
df3 
```

```python
df4 = df3 
```

```python
df4['embark_town'].fillna(df4['embark_town'].mode()[0],inplace=True)
```

```python
df4['embarked'].fillna(df4['embarked'].mode()[0])
```
