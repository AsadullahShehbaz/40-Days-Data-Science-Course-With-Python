```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
```

```python
df.drop(columns=['deck'], inplace=True)
```

```python
df.shape
```

```python
sns.countplot(x='survived', data=df,hue='survived' )
```

```python
sns.barplot(data=df,x='survived', y='age', hue='survived')
```

```python
sns.histplot(data=df, x='age', hue='survived', multiple='stack')
```

```python
df['age'].hist(bins=30, edgecolor='black')
```

```python
sns.boxplot(data=df,x='survived',y='age',hue='who')
```

```python
sns.scatterplot(data=df,x='age',y='fare')
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='pclass')
plt.ylim(0,270)
```

```python
sns.pairplot(df,hue='survived')
```

```python
df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
sns.swarmplot(data=df,x='survived',y='fare',hue='who')
```

```python
sns.catplot(x='pclass',y='fare',data=df,hue='embarked',kind='point')
```

```python
sns.lmplot(x='age',y=(df['fare']<300),data=df,hue='who')
```

```python
df['age_groups'] = pd.cut(
                        df['age'],
                          [0, 10, 20, 30,50,80],
                          labels=['Kids','Child','Youth','Middle Aged','Old']
                          )
```

```python
df.head()
```

```python
df.groupby(['age_groups','who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
sns.pairplot(df,vars=['age','fare','pclass'],hue='survived')
```
